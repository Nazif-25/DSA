```python
import sys

inversions = 0

def mergeSort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = mergeSort(arr[:mid])
    right = mergeSort(arr[mid:])
    return merge(left, right)

def merge(a, b):
    global inversions
    temp = []
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            temp.append(a[i])
            i += 1
        else:
            temp.append(b[j])
            inversions += len(a) - i  # Correct inversion count
            j += 1
    temp.extend(a[i:])
    temp.extend(b[j:])
    return temp

size = int(sys.stdin.readline())
arr = list(map(int, sys.stdin.readline().split()))
sorted_arr = mergeSort(arr)
print(inversions)
print(' '.join(map(str, sorted_arr)))
print(sorted_arr)
```

```python
MOD = 107
def find_modulus(a, b):
    if b == 0:
        return 1
    elif b % 2 != 0:
        return a * find_modulus(a, b-1) % MOD
    else:
        result = find_modulus(a, b/2)
        return result * result % MOD

a, b = map(int, input().split())
print(find_modulus(a, b))
```

```python
# Fast Matrix Drift
def multiply_matrices(A, B):
  return [
           [(A[0][0]*B[0][0] + A[0][1]*B[1][0]) % MOD, (A[0][0] * B[0][1] + A[0][1] * B[1][1]) % MOD],
           [(A[1][0]*B[0][0] + A[1][1]*B[1][0]) % MOD, (A[1][0] * B[0][1] + A[1][1] * B[1][1]) % MOD],
         ]


def matrix_power(matrix, power):
      result = [[1, 0], [0, 1]]
      while power > 0:
        #   checks if power or exponent is odd or even
            if power % 2 == 1:
                result = multiply_matrices(result, matrix)

            matrix = multiply_matrices(matrix, matrix)
            power //= 2
      return result


MOD = (10) ** 9 + 7
tests = int(input())
for i in range(tests):
  arr = [int(x) for x in input().split()]
  x = int(input())
  A = [[arr[0], arr[1]],
      [arr[2], arr[3]]]
  val = matrix_power(A, x)
  print(val[0][0], val[0][1])
  print(val[1][0], val[1][1])
```

```python
import sys


def build_bst(arr, low, high, result):
    if low > high:
        return result
    mid = (low+high)//2
    result.append(arr[mid])
    build_bst(arr, low, mid - 1, result)
    build_bst(arr, mid + 1, high, result)


size = int(input())
arr = list(map(int, sys.stdin.readline().split()))
result = []
build_bst(arr, 0, len(arr) - 1, result)
for _ in result:
    print(_, end=" ")
```
